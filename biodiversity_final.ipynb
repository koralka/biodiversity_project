### Statistical significance
This section will run some chi-squared tests to see if different species have statistically significant differences in conservation status rates. In order to run a chi squared test, a contingency table will need to be created. The contingency table should look like this:

||protected|not protected|
|-|-|-|
|Mammal|?|?|
|Bird|?|?|

The first test will be called `contingency1` and will need to be filled with the correct numbers for mammals and birds.

```python
contingency1 = [[30, 146],
               [75, 413]]
chi2_contingency(contingency1)
```


The results from the chi-squared test returns many values, the second value which is 0.69 is the p-value. The standard p-value to test statistical significance is 0.05. For the value retrieved from this test, the value of 0.69 is much larger than 0.05. **In the case of mammals and birds there doesn't seem to be any significant relationship between them i.e. the variables independent.**

The next pair, is going to test the difference between `Fish` and `Mammal`.

The format is again is like below:

||protected|not protected|
|-|-|-|
|Mammal|?|?|
|Fish|?|?|

```python
contingency2 = [[30, 146],
               [11, 115]]
chi2_contingency(contingency2)
```

The p-value is 0.056 - it is still a bit larger than 0.05. **There doesn't seem to be a significant relationship between mammals and fish either - the variable independent.**

The next pair, is going to test the difference between `Reptile` and `Mammal`.

```python
contingency2 = [[30, 146],
               [5, 73]]
chi2_contingency(contingency2)
```

This time the p-value is 0.039 which is below the standard threshold of 0.05 which can be take that the difference between reptile and mammal is statistically significant. **Mammals are shown to have a statistically significant higher rate of needed protection compared with Reptiles.**

### Species in Parks
Scientific names might not be the most informative for the majority of people. Next step is to have a look at the common names from `species` to find the most prevalent animals in the dataset.   
Data will need to be split up into individual names.

As a reminder, let's have a look at `species_df1` again 

```python
species_df1.head()
```

```python
# function to remove punctuation from 'common_names' column
def remove_punctuation(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

common_Names = species_df1[species_df1.category == "Mammal"]\
    .common_names\
    .apply(remove_punctuation)\
    .str.split().tolist()

common_Names[:10]
```

Next, let's clean up duplicate words in the new `common_Names` list.

```python
cleanRows = []

for item in common_Names:
    item = list(dict.fromkeys(item))
    cleanRows.append(item)

cleanRows[:10]
```

Creating one list of words:

```python
res = list(chain.from_iterable(i if isinstance(i, list) else [i] for i in cleanRows))
res[:10]
```

Let's count the number of occurrences of each word to find the most common common name.

```python
words_counted = []

for i in res:
    x = res.count(i)
    words_counted.append((i,x))

pd.DataFrame(set(words_counted), columns =['Word', 'Count']).sort_values("Count", ascending = False).head(20)
```

`Bat` occurred 23 times while `Shrew` came up 18 times.

Back to the scientific names - there are several different scientific names for different types of bats. Let's find out which rows of `species` refer to bats by adding a new `is_bat` column.

```python
species_df1['is_bat'] = species_df1.common_names.str.contains(r"\bBat\b", regex = True)

species_df1.head()
```

```python
species_df1[species_df1.is_bat]
```

**Yellowstone National Park** seems to have **the largest number** with **8,362 observations** and the Great Smoky Mountains National Park having the lowest with 2,411.
